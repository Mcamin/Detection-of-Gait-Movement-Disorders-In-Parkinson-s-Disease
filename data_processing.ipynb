```python
# Gets the relative path of the data files
def get_filepaths(directory):
    return glob.glob(directory, recursive=True)
```

```python
#  Save file to csv
def save_to_csv(data,output_path):
    data.to_csv('r'+output_path, sep='\t', encoding='utf-8', header='true')
```
