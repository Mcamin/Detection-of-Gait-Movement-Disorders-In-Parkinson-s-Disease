```python
# Libraries
import glob
import pandas as pd
import re
```

```python
# Global variables
data_dir = '.\Data\**\*.txt'
column_names= ['Id', 'Timestamp', 'EEG_FP1', 'EEG_FP2', 'EEG_F3', 'EEG_F4', 'EEG_C3', 'EEG_C4', 'EEG_P3', 'EEG_P4', 'EEG_O1', 'EEG_O2', 'EEG_F7', 'EEG_F8', 'EEG_P7', 'EEG_P8', 'EEG_FZ', 'EEG_CZ', 'EEG_PZ', 'EEG_FC1', 'EEG_FC2', 'EEG_CP1', 'EEG_CP2', 'EEG_FC5', 'EEG_FC6', 'EEG_CP5', 'EEG_CP6','EMG1_LTA','EMG2_RTA','IO','EMG3_ECG','EMG4_RGS','LShankACCX', 'LShankACCY', 'LShankACCZ', 'LShankGYROX', 'LShankGYROY','LShankGYROZ' ,'LShank_NC','RShankACCX', 'RShankACCY', 'RShankACCZ', 'RShankGYROX',
               'RShankGYROY','RShankGYROZ','RShank_NC','WaistACCX','WaistACCY','WaistACCZ','WaistGYROX','WaistGYROY','WaistGYROZ','Waist_NC','ArmACCX','ArmACCY','ArmACCZ','ArmGYROX','ArmGYROY','ArmGYROZ','SC','Label']
```

```python
# Gets the relative path of the data files
def get_filepaths(directory):
    return glob.glob(directory, recursive=True)
```

```python
#  Save file to csv
def save_to_csv(data,filename):
    data.to_csv('./Output/'+filename, sep=',', encoding='utf-8', header='true')
```

```python
def process_columns(df,new_col_names):
    df.rename(new_col_names, axis='columns', inplace =True)
```

```python
# Fix the column order and write the tasks in csv files
def process_data():
    for filepath in get_filepaths(data_dir):
        patient_nbr= int(re.findall(r'0\d{2}', filepath)[0])
        task_nbr= int(re.findall(r'\\task_(\d)', filepath)[0])
        trial_nbr = 1
        filename = str(patient_nbr)+"_trial_"+str(trial_nbr)+"_task_"+str(task_nbr)+".csv"
        print("Processing patient n°: %d, task n°: %d" % (patient_nbr, task_nbr))
        task =  pd.read_csv(filepath, sep=',', header=None, names=column_names)
        if patient_nbr in [1,2,6,7]:
            ### For 1,2,6,7,8-1 change column EMG1 <> EMG2
            new_cols_names = {"EMG1_LTA": "EMG2_RTA", "EMG2_RTA": "EMG1_LTA"}
            process_columns(task,new_cols_names)
        elif patient_nbr == 8:
            trial_nbr = int(re.findall(r'\\OFF_(\d)', filepath)[0])
            ### For 8-1 change column EMG1 <> EMG2
            new_cols_names = {"EMG1_LTA": "EMG2_RTA", "EMG2_RTA": "EMG1_LTA"}
            if trial_nbr == 2:
                ### For 8-2, change column EMG1 > EMG2, EMG2 > EMG4, EMG4 > EMG1
                new_cols_names = {"EMG1_LTA": "EMG2_RTA", "EMG2_RTA": "EMG4_RGS", "EMG4_RGS": "EMG1_LTA"}
            process_columns(task,new_cols_names)
        elif patient_nbr == 9:
            ### For 9, change column EMG3 <> EMG4
            new_cols_names = {"EMG3_ECG": "EMG4_RGS", "EMG4_RGS": "EMG3_ECG"}
            process_columns(task,new_cols_names)
        # Save df to csv files
        save_to_csv(task,filename)

```

```python
process_data()
```
